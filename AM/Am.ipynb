```python
filtered_df = df[
    (df['price'] < 100) &
    (df['room_type'] == 'Entire home/apt') &
    (df['neighbourhood_group'] == 'Manhattan')
]

print(filtered_df.head())
```

```python
def price_category(price):
    if price < 100:
        return "Low"
    elif price < 200:
        return "Medium"
    else:
        return "High"

df['price_category'] = df['price'].apply(price_category)

print(df[['price', 'price_category']].head())
```

```python
sorted_df = df.sort_values(by='price', ascending=False)

print(sorted_df[['name', 'price']].head())
```

Q1 — What is EDA? Why is it important?
EDA (Exploratory Data Analysis) is the process of understanding the dataset by using summaries, statistics, and visualizations before applying any model. It is important because it helps us find patterns, detect missing values, identify outliers, and understand relationships between variables. It basically gives a clear idea about the data so we can make better decisions.

Q3 — What insights can we get from describe()?
The describe() function gives statistical summary of numerical columns. From this, we can see mean (average), min and max values, standard deviation (spread of data), and quartiles. It helps us quickly understand how the data is distributed and whether there are any extreme values or not.
