(because it uses labeled data: spam / not spam)
A retail company groups customers based on purchasing behavior without any predefined labels
→ Unsupervised Learning
(because there are no labels, just grouping patterns)

(because it learns using rewards and trial-and-error)
A model predicts house prices based on features like area, location, and number of rooms
→ Supervised Learning

Regression vs Classification

Regression
- Used when the target variable is continuous (numeric)
- Output is a number
- Example: predicting house price, salary
Classification
- Used when the target variable is categorical
- Output is a class/label
- Example: spam/not spam, pass/fail

| Target Variable Type                     | ML Type        | Justification                |
| ---------------------------------------- | -------------- | ---------------------------- |
| Numeric (e.g., price, age)               | Regression     | Because output is continuous |
| Categorical (e.g., yes/no, class labels) | Classification | Because output is discrete   |


````python
import pandas as pd

data = {
    'Area': [1000, 1500, 2000, 1200, 1800],
    'Bedrooms': [2, 3, 4, 2, 3],
    'Location_Score': [7, 8, 9, 6, 8],
    'Price': [50, 75, 100, 60, 85]   
}

df = pd.DataFrame(data)
print(df)
````

````python
X = df[['Area', 'Bedrooms', 'Location_Score']] 
y = df['Price']                                

print("Features (X):")
print(X)

print("\nTarget (y):")
print(y)
````

#  Regression vs Classification Examples

##  1. Regression Example

### Example: House Price Prediction

- **Input:**  
  Area, number of bedrooms, location score  

- **Output:**  
  House price (numeric value)  

- **Use Case:**  
  Real estate companies use this to estimate property prices based on features  

**Justification:**  
This is a regression problem because the output (price) is a continuous numeric value.

---

##  2. Classification Example

### Example: Email Spam Detection

- **Input:**  
  Email text, keywords, sender details  

- **Output:**  
  Spam or Not Spam (category)  

- **Use Case:**  
  Email services use this to filter unwanted emails automatically  

**Justification:**  
This is a classification problem because the output is a category (label), not a number.

---

##  Summary

- **Regression →** predicts continuous numerical values  
- **Classification →** predicts categorical labels  



##  Q1 — What are types of machine learning?

There are mainly three types of machine learning:

- **Supervised Learning**  
  The model learns from labeled data (input → output).  
  Example: predicting house prices, spam detection  

- **Unsupervised Learning**  
  The model finds patterns in data without labels.  
  Example: customer segmentation  

- **Reinforcement Learning**  
  The model learns by trial and error using rewards and penalties.  
  Example: robot learning, game playing
